```python
import random
from datasets import Dataset, Audio
import evaluate
from transformers import WhisperProcessor, WhisperForConditionalGeneration, WhisperFeatureExtractor, WhisperTokenizer, TrainingArguments, Trainer, Seq2SeqTrainer, Seq2SeqTrainingArguments
import torch
import torchaudio
```

```python
# Load CSV and sample
df = pd.read_csv("/kaggle/input/the-uyghur-voice-cup/train.csv")
df['audio_path'] = "/kaggle/input/the-uyghur-voice-cup/" + df['filepath']
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
train_df = df.iloc[:6000]
# train_df=df
# val_df = df.iloc[500:600]

# Convert to Hugging Face Datasets
train_ds = Dataset.from_pandas(train_df).cast_column("audio_path", Audio(sampling_rate=16000))
# val_ds = Dataset.from_pandas(val_df).cast_column("audio_path", Audio(sampling_rate=16000))
```

```python
# Load processor & model
checkpoint = "ixxan/whisper-small-uyghur-thugy20"
processor = WhisperProcessor.from_pretrained(checkpoint)
model = WhisperForConditionalGeneration.from_pretrained(checkpoint)
```

```python
# Freeze all but the final layers
# for name, param in model.model.named_parameters():
#     if not any(layer in name for layer in ["layers.11", "encoder.positional_embedding", "decoder.positional_embedding"]):
#         param.requires_grad = False

def prepare_example(batch):
    # Extract audio array
    audio = batch["audio_path"]["array"]
    
    # Prepare input features
    inputs = processor(audio, sampling_rate=16000, return_tensors="pt")
    
    # Manually tokenize the target text
    with processor.tokenizer.as_target_tokenizer():
        labels = processor.tokenizer(batch["transcription"], return_tensors="pt").input_ids

    # Add to batch
    batch["input_features"] = inputs.input_features[0]
    batch["labels"] = labels[0]
    return batch

train_ds = train_ds.map(prepare_example)
# val_ds = val_ds.map(prepare_example)
```

```python
# Training args
training_args = Seq2SeqTrainingArguments(
    output_dir="./whisper-ft-uyghur",
    per_device_train_batch_size=4,
    # per_device_eval_batch_size=2,
    gradient_accumulation_steps=2,
    learning_rate=1e-5,
    num_train_epochs=3,
    # logging_steps=100,
    # eval_strategy="epoch",
    # save_strategy="epoch",
    fp16=torch.cuda.is_available(),
    predict_with_generate=True,
    report_to="none",
    dataloader_num_workers=4,
    max_steps=1000
)

# Data collator
def data_collator(batch):
    input_features = [example["input_features"] for example in batch]
    labels = [example["labels"] for example in batch]
    input_features = torch.tensor(input_features)
    labels = processor.tokenizer.pad({"input_ids": labels}, return_tensors="pt")["input_ids"]
    labels[labels == processor.tokenizer.pad_token_id] = -100
    return {"input_features": input_features, "labels": labels}

# Metric: CER
cer_metric = evaluate.load("cer")

def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)
    return {"cer": cer_metric.compute(predictions=pred_str, references=label_str)}

# Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    # eval_dataset=val_ds,
    data_collator=data_collator,
    tokenizer=processor.feature_extractor,
    compute_metrics=compute_metrics
)

#  Fine-tune
trainer.train()
```

## submission on fine tuned model

```python
import torch
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq
import torchaudio
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import re
import unicodedata
```

```python
# Settings
beam_size = 3
temperature = 0.3
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Normalization (used only to clean output text)
def normalize_text(text):
    text = unicodedata.normalize("NFKD", text)
    text = "".join(c for c in text if not unicodedata.combining(c))
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Load audio
def load_audio(path):
    audio, sr = torchaudio.load(path)
    if sr != 16000:
        audio = torchaudio.transforms.Resample(sr, 16000)(audio)
    return audio[0]

# Transcribe function
def transcribe(audio_path):
    audio = load_audio(audio_path)
    inputs = processor(audio, sampling_rate=16000, return_tensors="pt").to(device)

    with torch.no_grad():
        generated_ids = model.generate(
            inputs["input_features"],
            do_sample=True,
            temperature=temperature,
            num_beams=beam_size,
            max_length=448,
        )

    pred = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return normalize_text(pred)

# Load test.csv
test_df = pd.read_csv("/kaggle/input/the-uyghur-voice-cup/test.csv")

# Run transcription
results = []
for _, row in tqdm(test_df.iterrows(), total=len(test_df)):
    audio_path = Path("/kaggle/input/the-uyghur-voice-cup") / row["filepath"]
    pred = transcribe(str(audio_path))
    results.append((row["ID"], pred))

# Save submission
submission_df = pd.DataFrame(results, columns=["ID", "transcription"])
submission_df.to_csv("submission.csv", index=False)

print(" submission.csv saved!")
```
